## drop
used to drop entire row or column  
__Note__: you need to specify the axis  
__column__: axis = 1  
__Row__: axis = 0  
__inplace__ is used to just assign it back to the DataFrame  

```python
first.drop("Wheels",axis=1,inplace=True)
# we can also use normal assignment and remove 'inplace' attribute
```

```python
first
```
