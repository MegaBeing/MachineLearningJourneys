### For add_axes()

## Matplotlib Workflow

```python
# 1. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
# 2. prepare data
x=[1,2,3]
y=[1,2,3]
# 3. setup the plot
fig,axe = plt.subplots(figsize=(10,10)) #(width, height)
# 4. plot data
axe.plot(x,y)
# 5. customize plot
axe.set(title="Simple plot",xlabel="X-Axis",ylabel="Y-Axis")
# 6. save (Just save the whole figure in a file)
fig.savefig("sample-plot.jpeg")
```

### creating plots straight away from the dictionary

```python
butter = {"Peanut butter":10,"Normal butter":8,"Cashew butter":12}
fig,axe = plt.subplots()
axe.bar(butter.keys(),butter.values()) # creating bar graph(default is vertical) 
axe.set(title="Prices",xlabel="Butter",ylabel="Dollars")# title=title of the plot, xlabel = label on x-axis, ylabel= label on y-axis
fig.savefig("Butter_prices.jpeg")
```

### Horizontal bar graph

```python
fig, ax = plt.subplots()
ax.barh(list(butter.keys()),list(butter.values()))
```

### Histogram

```python
x=np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

### Subplots
#### option 1: (RECOMMENDED)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(20,10))
ax1.plot(x,x/2);
ax2.scatter(x,x**2)
ax3.bar(butter.keys(),butter.values())
ax4.barh(list(butter.keys()),list(butter.values()))
```

### Object oriented Matplotlib

##### when to use pyplot and oo plot?  
* when you want to plot something quickly, we use pyplot
* when you want to plot something more advance, we use oo plot

```python
over_50=heart[heart["age"]>50]
over_50
```

```python
# pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target',figsize=(10,10));
```

```python
# OO METHOD mixed with pyplot
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax)
ax.set(xlim=[45,100])# we are changing the limits of the plot from 45 to 100
```

as we can see oo method is better as we have a lot more control on the figure we are going to display thus making it an ideal method in data analysis

```python
# OO FROM SCRATCH
fig,ax = plt.subplots(figsize=(10,10))
# PLOT THE DATA
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#customize the plot
ax.set(title='Age v Cholestrol',
       xlabel='Age',
       ylabel='Cholestrol')

# add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# add a horizontal line
ax.axhline(over_50.chol.mean(),linestyle="--",c='red');
```

```python
fig,ax=plt.subplots(figsize=(10,10))
scatter = ax.scatter(x=over_50.chol,y=over_50.thalach,c=over_50.oldpeak)
ax.set(title="Cholestrol vs Max HeartRate",xlabel="cholestrol",ylabel="Max HeartRate")
ax.legend(*scatter.legend_elements(),title="old peak")
ax.axhline(y=over_50.thalach.mean(),linestyle="--",c='red')
```

```python
faltu = np.random.randint(0,10,size=(10,4))
df=pd.DataFrame(faltu,columns=['a','b','c','d'])
df # one random dataframe created to plot the data for an example
```

```python
# costumizing the plots
fig,ax=plt.subplots(figsize=(10,10))
scatter=ax.scatter(x=df.a,y=df.index,cmap='winter',c=df.d)
ax.set(title="Random Data",xlabel='a',ylabel='Values')
ax.legend(*scatter.legend_elements(),title='d')
fig.suptitle('Rand')# super title on the figure
```
